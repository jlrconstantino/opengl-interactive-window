# Eventos de teclado e de mouse

```python
# Transformador da nave espacial
class NaveTransformer:

    # Inicialização
    def __init__(self, nave):
        self.tx = 0.0
        self.ty = 0.0
        self._nave = nave
        self.center = np.array(nave.center)
    
    # Transformação da nave
    def _transform(self):
        self._nave.transformation = translate3D(self.tx, self.ty)

    # Mover para cima
    def move_up(self):
        def closure(*_):
            nonlocal self
            self.ty += 0.02
            self.center[1] += 0.02
            self._transform()
        return closure

    # Mover para direita
    def move_right(self):
        def closure(*_):
            nonlocal self
            self.tx += 0.02
            self.center[0] += 0.02
            self._transform()
        return closure

    # Mover para baixo
    def move_down(self):
        def closure(*_):
            nonlocal self
            self.ty -= 0.02
            self.center[1] -= 0.02
            self._transform()
        return closure
    
    # Mover para esquerda
    def move_left(self):
        def closure(*_):
            nonlocal self
            self.tx -= 0.02
            self.center[0] -= 0.02
            self._transform()
        return closure

# Instanciação do transformador da nave
nave_transformer = NaveTransformer(nave)

# Transformação dos animais
def transform_animal(animal, nave_transformer):

    # Parâmetros
    tx = 0.0
    ty = 0.0
    sx = 1.0
    sy = 1.0
    t = 0.0
    center = animal.center

    # Closure
    def closure(*args):
        nonlocal tx, ty, sx, sy, t, animal, center, nave_transformer

        # Para atualizar os parâmetros
        dist = nave_transformer.center - center

        # Deslocamento relativo
        tx += dist[0]/32
        ty += dist[1]/32
        trans = translate3D(tx, ty)

        # Escalonamento
        sx -= 0.06
        if sx < 0.0:
            sx = 0.0
        sy -= 0.06
        if sy < 0.0:
            sy = 0.0
        scale = scale3D(sx, sy)

        # Rotação
        t += np.pi/8
        rotat = relative_zrotate3D(t, x0=center[0], y0=center[1])

        # Transformação
        animal.transformation = np.linalg.multi_dot((trans, scale, rotat))
        
    return closure


# Transformação do moinho
def transform_moinho(moinho):
    
    # Parâmetros
    t = 0.0
    def closure(*args):

        # Atualização dos parâmetros
        nonlocal t, moinho
        t += np.pi/8

        # Matriz
        rotat = relative_zrotate3D(t, x0 = -0.54, y0 = 0.56)

        # Transformação
        moinho.transformation = rotat
        
    return closure

# Atribuição dos eventos de tecla
window.add_key_callback("M", transform_animal(mimosa, nave_transformer))
window.add_key_callback("C", transform_animal(caramelo, nave_transformer))
window.add_key_callback("K", transform_animal(klovis, nave_transformer))
window.add_key_callback("V", transform_moinho(moinho))
window.add_key_callback("W", nave_transformer.move_up())
window.add_key_callback("D", nave_transformer.move_right())
window.add_key_callback("S", nave_transformer.move_down())
window.add_key_callback("A", nave_transformer.move_left())
```
